```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
import seaborn as sns
```

```python
df=pd.read_csv("GlobalLandTemperaturesByCity.csv")
```

```python
data=df[df['City']=='New York'].copy()
data['dt']=pd.to_datetime(data['dt'],errors="coerce")
data['Year']=data['dt'].dt.year
annual=data.groupby('Year')['AverageTemperature'].mean().reset_index()
pt.figure(figsize=(12,6))
pt.plot(annual['Year'],annual['AverageTemperature'],linestyle="-",color="skyblue",marker=".")
pt.title("average temperature",fontsize=18)
pt.xlabel("Year",fontsize=16)
pt.ylabel("avg temp",fontsize=16)
pt.grid(True)
pt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
annual = df.groupby('Country')['AverageTemperature'].mean().reset_index()
annual_sort = annual.sort_values(by='AverageTemperature', ascending=False).head(50)

plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='AverageTemperature', data=annual_sort, palette="Blues")
plt.title("Average Temperature by Country", fontsize=18)
plt.xlabel("Country", fontsize=16)  
plt.ylabel("Avg Temperature (C)", fontsize=16)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
temperature_data=df['AverageTemperature'].dropna()
pt.figure(figsize=(12, 6))
pt.hist(temperature_data,bins=25,color='green',edgecolor='black',alpha=0.7)
pt.title("dist. temp",fontsize=18)
plt.xlabel("temperature ranges", fontsize=16)  
plt.ylabel("frequency", fontsize=16)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Drop rows with missing values in relevant columns
data = df[['AverageTemperature', 'AverageTemperatureUncertainty']].dropna()

# Plot scatter plot
plt.figure(figsize=(12, 6))
plt.scatter(data['AverageTemperature'], data['AverageTemperatureUncertainty'], alpha=0.6, color='blue', edgecolor='black')
plt.title('Correlation Between Average Temperature and Temperature Uncertainty', fontsize=18)
plt.xlabel('Average Temperature', fontsize=14)
plt.ylabel('Temperature Uncertainty', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python

import matplotlib.pyplot as plt

# Filter the dataset for the specific country
country_data = df[df['Country'] == 'India'].dropna()

# Extract the relevant data: Year and AverageTemperature
country_data['Year'] = pd.to_datetime(country_data['dt']).dt.year.head(100)
temperature_data = country_data[['Year', 'AverageTemperature']]

# Plot boxplot for temperature spread by year
plt.figure(figsize=(12, 6))
plt.boxplot(
    [temperature_data[temperature_data['Year'] == year]['AverageTemperature'] for year in temperature_data['Year'].unique()],
    labels=temperature_data['Year'].unique(),
    patch_artist=True,
    boxprops=dict(facecolor='purple', color='black'),
    medianprops=dict(color='yellow', linewidth=1.5),
    flierprops=dict(marker='o', color='red', alpha=0.5),
)

plt.title('Spread of Average Temperatures in the India (by Year)', fontsize=18)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Average Temperature', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import seaborn as sns
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.histplot(data=tips,x='total_bill',kde=True,bins=20,color='skyblue')
pt.title("Disturbution")
pt.show()
```

```python
sns.histplot(data=tips,x='day',y='total_bill',kde=True,palette='viridis')
pt.title("Disturbution")
pt.show()
```

```python
sns.boxplot(data=tips,x='total_bill',hue='sex',palette='viridis')
pt.title("box plot")
pt.show()
```

```python
sns.scatterplot(data=tips,x='total_bill',y="tip",hue='day',style="time",palette='deep')
pt.title("scatter plot")
pt.show()
```

```python
sns.lineplot(data=tips,x='size',y='tip',marker='*',color='orange')
pt.title("line plot")
pt.show()
```

```python
sns.pairplot(data=tips,hue='sex',palette='husl')
pt.title("pair plot")
pt.show()
```

```python
n=tips.select_dtypes(include='number')
correlation=n.corr()
sns.heatmap(correlation,annot=True,cmap='coolwarm',fmt='.2f')
pt.show()
```
